```python
# there are 88 neighborhoods
# for each month, there are 8 reports for each nbhd, totaling 704 unique entries
# there are 23 months reported
Total_bins = len(crime_df)/704

bins = np.arange(0,16192+704, 704)
binlabels = np.arange(1,24,1) #must be one fewer than bins
binlabels
```

```python
crime_df['Period'] = pd.cut(crime_df["OBJECTID"], bins, labels=binlabels)
crime_df
```

```python
crime_df.loc[(crime_df["Period"] == 1) & (crime_df["ucrDescription"] == "Aggravated Assault"), "number"].sum()
```
