```python
import gmaps
import pandas as pd

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create scooter dataframe
scooter_df = pd.read_csv('Downloads/scooter_address_lat_long.csv')
scooter_df.dropna()
scooter_df.head()
```

```python
# Create scooter dataframe
crime_df = pd.read_csv('Downloads/NEIGHBORHOOD_CRIME_STATS_Lat_Long.csv')
crime_df.dropna()

crime_df=crime_df.fillna(value=0)
crime_df['neighborhood']=crime_df['neighborhood'].str.upper()

crime_2018=crime_df.loc[crime_df['reportYear']==2018.0]

crime_2018.head()
```

```python
nstats=crime_2018.groupby('neighborhood').agg({'number':'sum','Lat':'first','Long':'first'})

nstats.reset_index(inplace=True)

nstats
```

```python
locations2 = nstats[['Lat','Long']].astype(float)

crime_rate = nstats['number'].astype(float)
```

```python
#Heat layer

heat_layer = gmaps.heatmap_layer(locations2, weights=crime_rate,
                                dissipating=False,max_intensity=100,
                                point_radius = 1)


heat_layer.dissipating = False
heat_layer.max_intensity = 1000
heat_layer.point_radius = 1


```

```python
# Store Lat and Long in locations
locations = scooter_df[["Lat", "Long"]].astype(float)
```

```python
# Create bank symbol layer
location_layer = gmaps.symbol_layer(locations, fill_color="blue",
                                stroke_color="blue", scale=2)
```

```python
# Plot Map
fig = gmaps.figure()
fig.add_layer(location_layer)
fig.add_layer(heat_layer)
fig
```
